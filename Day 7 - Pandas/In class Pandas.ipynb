```python
#series obj - single column, 1D column
temp11 = pd.Series([32,30,31,29,30])
print("Type of temp 11: " ,type(temp11))

#a dictionary of temperatures
#python data object
city_temp = {
    'City' : ["Kuala Lumpur", "Banting", "Klang", "Shah Alam", "Damansara"],
    'temp10' : [32, 29, 31, 30, 28]
}

#creating data frame
temp_df = pd.DataFrame(city_temp)
print("Type of temp_df: " ,type(temp_df))

print(temp_df) #df obj
print(temp11)   #series obj
```

```python
df2 = temp_df #dataframe

df2["temp11"] = temp11

df2
```
