```python
#involve multiple column = dataframe
species2 = df.drop(['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], axis = 1)
species2.head()
type(species2)
```

```python
#calculate column
#series 
sepal_area = iris_without_species['sepal_length'] * iris_without_species['sepal_width']
sepal_area.head()
```

```python
iris_without_species['sepal_area'] = sepal_area
iris_without_species.head()
```

```python
petal_area = iris_without_species['petal_length'] * iris_without_species['petal_width']
petal_area.head()
```

```python
iris_without_species['petal_area'] = petal_area
iris_without_species.head(50)
```

Filtered out parts of dataframes using pandas

```python
#only part of iris_wout_species value of sepal_area is > 14
#need to write targeted column
largeFlower = iris_without_species[iris_without_species['sepal_area'] > 15]
largeFlower.head(10)
```

```python
#how many large flowers we heave in iris dataset
#how many flowers that has sepal_area of exactly =16
largeFlower.info()
```

```python
flower16 = iris_without_species[iris_without_species['sepal_area'] == 16]
flower16.head(10)
#print("Number of flowers has exactly sepale_area equals to 16 : ", len(flower16))
```

```python
#sorting values 
sortedLargeFlowers = largeFlower.sort_values(by='sepal_area', ascending=False)
sortedLargeFlowers
```

```python
largestFlower = sortedLargeFlowers['sepal_area']
largestFlower.head(1)
```

```python
areaLargestFlower = sortedLargeFlowers['sepal_area'].max()
areaLargestFlower
```
