```python
students = ['Ali', 'Meena', 'Naim','Syed','Nur','Abu']
grades = [98, 75, 95, 86, 42, 90]
```

```python
import matplotlib.pyplot as plt

#create a bar plot
plt.bar(students, grades, color = 'green') #specify properties of our bar cgart
plt.title('Chart of Students Grades')
plt.xlabel('Students Names')
plt.ylabel('Marks')
plt.show() #if bar not appear, use this
```

```python
#scatterplot

plt.scatter(students, grades, color = 'green') #specify properties of scatterplots

plt.title('Chart of Students Grades')
plt.xlabel('Students Names')
plt.ylabel('Marks')
plt.show() #if bar not appear, use this
```

```python
#line plot

plt.plot(students, grades, color = 'green') #specify properties of lineplots

plt.title('Chart of Students Grades')
plt.xlabel('Students Names')
plt.ylabel('Marks')
plt.show() #if bar not appear, use this
```

```python
#all of them in same place

#create a figure of 15 row and 5 columns
plt.figure(figsize = (15,5)) #lagi kecik size, cepat pecah bila zoom

#subplot
plt.subplot(1,3,1) #rows, columns position
plt.bar(students, grades, color = 'green')

#subplot
plt.subplot(1,3,2) #rows, columns 2nd
plt.plot(students, grades, color = 'green')

#subplot
plt.subplot(1,3,3) #rows, columns 3rd
plt.scatter(students, grades, color = 'green')
```
